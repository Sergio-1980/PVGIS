# 1- librerias 

```python
import warnings
warnings.simplefilter('ignore')

import pandas as pd
pd.set_option('display.max_columns',None)

import numpy as np
```

# 2- Extraccion de datos

```python
df = pd.read_html("Registro_15_07_2021.xls", flavor='html5lib')
df = df[0]
df.columns = df.iloc[0]
df=df.drop(0)
df = df.reset_index(drop=True)
df
```

```python
df.columns=[e.lower().replace(' ','_') for e in df.columns]
df.columns
```

# 3-Analaisis exploratorio

### 3.1- Analisis descriptivo

```python
df.describe()
```

### 3.2- Analisis por estado  


```python
estado_df=df.groupby('estado').count().clave_de_registro
estado_df
```

```python
estado_df/df.shape[0]
```

```python
ax = estado_df.plot(kind='bar',color=['g','r'],grid=True,figsize=(9,6),title='Estado')
ax.set_xlabel('estado')
ax.set_ylabel('#registros');
```

**Se observa con mas de un 93% que el estado es defintivo**
